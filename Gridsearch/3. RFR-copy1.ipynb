```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dep=dataset[['charges']]
```

```python
#remove train and testset split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep, dep, test_size=1/3, random_state=0)
```

```python
#remove standardisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#model creation with Cross Validation and Grid Search
from sklearn.model_selection import GridSearchCV

param_grid={'criterion':['squared_error','absolute_error','friedman_mse'], 'max_features':['auto','sqrt','log2'], 'n_estimators':[10,100]}

grid=GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)

#fitting the model for grid search
grid.fit(indep, dep)
```

```python
#print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_

print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
